```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2 as cv
```

```python
class Image:
    
    def __init__(self, path):
        self.array = mpimg.imread(path)
        self.R = self.array[:, :, 0]
        self.G = self.array[:, :, 1]
        self.B = self.array[:, :, 2]
        self.fft2d = np.fft.fft2(self.array)
        self.lfs = 10 * np.log(np.abs(self.fft2d) + 1)
        
    def grayscale(self):
        
        grayscale_image = np.zeros((self.array.shape[0], self.array.shape[1], 1))
        for i in range(0, self.array.shape[0]):
            for j in range(0, self.array.shape[1]):
                sumof_values = self.array[i,j].sum()
                grayscale_image[i, j] += sumof_values
        self.grayscalee = grayscale_image 
        plt.imshow(grayscale_image)
        
    def fft2d(self, array):
        
        channel = np.fft.fft2d(array)
        return channel
        
    def lfs(self, fft_array):
        
        lfs = 10 * np.log(np.abs(fft_array) + 1)
        return lfs
    
    def plot(self, array=1):
    
        if isinstance(array, int):
            plt.imshow(self.array)
        else:
            plt.imshow(array)
```

```python
image = Image("Alex1.png")
```

```python
image.plot(image.R)
```
